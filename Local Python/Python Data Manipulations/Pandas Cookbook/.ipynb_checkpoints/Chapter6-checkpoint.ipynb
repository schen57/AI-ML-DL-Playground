# 6.2 Use resampling to find the snowiest month
If we wanted the median temperature each month, we could use resample() method such as this:

```python
import numpy as np
plt.figure(figsize=(20,10))
weather_2012['Temp (C)'].resample('M').apply(np.median).plot(kind="bar")
plt.grid()
# What does resample do? I am guessing the "M" is the month? 
# Convenience method for frequency conversion and resampling of time series. 
# Object must have a datetime-like index
# It turns out that the resampling is a way to aggregate time series
```

```python
is_snowing.astype(float)[:10]
```

```python
# use resample to find the percentage of time it was snowing each month
is_snowing.astype(float).resample("M").apply(np.mean)
# Applying mean to boolean values directly would result in mean of the aggregated 
# Values
```

```python
plt.figure(figsize=(20,10))
is_snowing.astype(float).resample("M").apply(np.mean).plot(kind="bar");
plt.grid()
```

# 6.3 Plotting temperature and snowiness 


```python
weather_2012.head()
```

```python
temperature=weather_2012['Temp (C)'].resample("M").apply(np.median)
is_snowing=weather_2012["Weather"].str.contains("Snow")
snowiness=is_snowing.astype(float).resample("M").apply(np.mean)

# Name the columns
temperature.name="Temperature"
snowiness.name='Snowiness'
```

```python
stats=pd.concat([temperature,snowiness],axis=1) # Concat horizontally
stats
```

```python

stats.plot(kind="bar",subplots=True,figsize=(20,10))
plt.show();
```
