# 7.3 Fixing the nan values and string/float confusions
We can pass a na_values option to pd.read to clean this up. We need to specify that the column data type is string not float

```python
na_values=["NO CLUE","N/A","0"]
# Storing the na values first
requests=pd.read_csv("/Users/shengyuchen/Dropbox/Engagement - Business/My Hub/AI:ML:DL Playground/Local Python/Python Data Manipulations/Pandas Cookbook/311-service-requests.csv",na_values=na_values,dtype={"Incident Zip":str})
```
